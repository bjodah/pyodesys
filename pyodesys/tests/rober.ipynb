```python
from pyodesys.tests._robertson import get_ode_exprs
import sympy as sp
from sympy import symbols
sp.init_printing()
```

```python
fj = [get_ode_exprs(reduced=reduced) for reduced in range(4)]
```

```python
t = symbols('t')
y = A, B, C = symbols('A B C')
inits = symbols('A0 B0 C0')
p = symbols('k1 k2 k3')
p1 = p2 = p3 = p + inits
y1, y2, y3 = [B, C], [A, C], [A, B]
```

```python
f0, j0 = fj[0][0](t, y, p), fj[0][1](t, y, p)
f0, sp.Matrix(j0)
```

```python
f1, j1 = fj[1][0](t, y1, p1), fj[1][1](t, y1, p1)
f1, sp.Matrix(j1)
```

```python
f2, j2 = fj[2][0](t, y2, p2), fj[2][1](t, y2, p2)
f2, sp.Matrix(j2)
```

```python
f3, j3 = fj[3][0](t, y3, p3), fj[3][1](t, y3, p3)
f3, sp.Matrix(j3)
```

```python
diff1 = sp.Matrix(f1).jacobian(y1) - sp.Matrix(j1)
diff1.simplify()
diff1
```

```python
diff2 = sp.Matrix(f2).jacobian(y2) - sp.Matrix(j2)
diff2.simplify()
diff2
```

```python
diff3 = sp.Matrix(f3).jacobian(y3) - sp.Matrix(j3)
diff3.simplify()
diff3
```
